## Investigation Overview

My main goal is to investigate the impact of the parents on the score of their children. I'm also interested in knowing if native students score better than the other to figure out what can be the cultural impact.

## Dataset Overview

The data set contains 635 features and 485490 student survey results. It's a very reach dataset which contains more features than necessary in regards to the main interest of this presentation.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from matplotlib.gridspec import GridSpec

%matplotlib inline

sb.set(style='white')

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# Load
pisa_clean = pd.read_csv("pisa2012_clean.csv", encoding='latin-1');

# Time withe parent as int
pisa_clean.loc[:,'Study Time With Parent'] = pisa_clean['Study Time With Parent'].astype('int64')

# Parents Presence
parents_presence_dtype = pd.api.types.CategoricalDtype(categories=['Both','Single'])
pisa_clean.loc[:,'Parents Presence'] = pisa_clean['Parents Presence'].astype(parents_presence_dtype)

# Gender as category
gender_dtype = pd.api.types.CategoricalDtype(categories=['Female','Male'])
pisa_clean.loc[:,'Gender'] = pisa_clean.Gender.astype(gender_dtype)

# Parent highest level of education as category
parentHE_dtype = pd.api.types.CategoricalDtype(
                    categories=['None','ISCED 1','ISCED 2','ISCED 3B, C','ISCED 3A, ISCED 4','ISCED 5B','ISCED 5A, 6'],
                    ordered=True)
pisa_clean.loc[:,'Parents Highest Education'] = pisa_clean['Parents Highest Education'].astype(parentHE_dtype)

# Immigration as category
immig_dtype = pd.api.types.CategoricalDtype(categories=['Native','Second-Generation','First-Generation'], ordered=True)
pisa_clean.loc[:,'Immigration Status'] = pisa_clean['Immigration Status'].astype(immig_dtype)
```

## Cultural impact : Native and non native students

11.5% of the students are not native. Their scores are slighly less than for native but the difference is very low (less than 10 points).

In Math, scores are similar between native and second generation students while the one for first generation is a bit less. This may be due to the abstract nature of mathematics (where the literary matters less). Afterall mathematics is oftenly quoted as the universal language.

```python
fig1, ax1 = plt.subplots(1, 1, figsize=(10, 10), sharex=True)
patches, texts, autotexts = ax1.pie(pisa_clean['Immigration Status'].value_counts(), 
                                    colors = ['#4c516d','#0080ff','#008ecc'],
                                    labels=['Native', 'Second Generation', 'First Generation'], 
                                    autopct='%1.1f%%',
                                    startangle=180
                                   );
for text in texts:
    text.set_color('grey')
    
for text in autotexts:
    text.set_color('white')
    
ax1.axis('equal');
plt.tight_layout();
plt.suptitle('How many native or non native students participate to the survey?')
plt.show();
```

```python
data = pd.melt(pisa_clean, id_vars=['Student Id', 'Gender', 'Country', 'Study Time With Parent', 'Parents Highest Education', 'Immigration Status', 'Parents Presence'], 
               value_vars=['Overall Score','Math Score','Reading Score','Science Score'], 
               var_name="Score Type", value_name='Score')
data['Score Type'] = data['Score Type'].replace('Overall Score', 'Overall').replace('Math Score','Math').replace('Reading Score','Reading').replace('Science Score','Science')

plt.figure(figsize=[14, 10]);
ax = sb.boxplot(x = data['Score Type'], y = data['Score'], hue=data['Immigration Status'], palette = sb.color_palette('husl', 8))
ax.xaxis.labelpad = 20
ax.yaxis.labelpad = 10
plt.suptitle('Is studying harder when you\'re not native?');
y_ticks = np.arange(0, 1000 + 1, 100)
plt.yticks(y_ticks, y_ticks);
```

## Does education of the parents matters when it comes to the education of the children?


We find out that in the dataset most of the parents reach the upper secondary education level. 
The first stage of tertiary education is the most common level of education, beside this no parent reach the second stage of the tertiary education.

The trends show that the highest the parent education is, the better the score is.

The upper secondary education ('ISCED 3B, C') is the inflection point. Before that point we observe the biggest score improvement. From that point, the students tend to perform above the scores mean.

```python
plt.figure(figsize=[14, 10]);
ax = sb.countplot(x = 'Parents Highest Education', data = pisa_clean, color = sb.color_palette()[0]);

# add annotations
n_points = pisa_clean.shape[0]
cat_counts = pisa_clean['Parents Highest Education'].value_counts()
locs, labels = plt.xticks() # get the current tick locations and labels

# loop through each pair of locations and labels
for loc, label in zip(locs, labels):

    # get the text property for the label to get the correct count
    count = cat_counts[label.get_text()]
    pct_string = '{:0.1f}%'.format(100*count/n_points)

    # print the annotation just below the top of the bar
    plt.text(loc, count-8, pct_string, ha = 'center', va='bottom', color = 'black')

plt.ylabel('Nr. of parents');
plt.xlabel('');
ax.yaxis.labelpad = 10
plt.suptitle('What\'s the education of the parents?', y = 1.02);
```

```python
parents_he_scores = pisa_clean[['Parents Highest Education','Overall Score','Math Score','Reading Score','Science Score']]
parents_he_scores.rename({'Overall Score':'Overall', 'Math Score':'Math', 'Reading Score':'Reading', 'Science Score':'Science'},axis=1, inplace=True)
data = pd.melt(parents_he_scores, id_vars=['Parents Highest Education'], 
               value_vars=['Overall','Math','Reading','Science'], 
               var_name="Score Type", value_name='Score Value')

fig, ax = plt.subplots(1, 1, figsize=(14, 10))

ax = sb.pointplot(ax=ax, data = data, 
                  x = 'Parents Highest Education', y='Score Value', hue='Score Type', 
                  dodge=0.3, palette=sb.color_palette('husl', 8), 
                  markers=["o", "x","x","x"], linestyles=["-", "--", "--", "--"]);
#ax = sb.pointplot(ax=ax, data = pisa_clean, x = 'Parents Highest Education', y='Overall Score', hue='Parents Presence', markers=["o", "x","x","x"], linestyles=["-", "--", "--", "--"]);
ax.set(xlabel='Parent Highest Education', ylabel='Score')
ax.xaxis.labelpad = 20
ax.yaxis.labelpad = 10
plt.suptitle("Do better educated parent(s) get smarter kids?", y=1.02);
```

## How much personal time should parent invest on the education of their children?

First the majority of the students don't study with their parents.

For those who do, some work up to 30 hours but most of them spend between 1 and 5 hours studying with their parents. Within that range we observe a drop of 70 points in the score so studying with parents does not seem to be a factor of success. Beyond 5 hours, the output is erratic and we can't draw further conclusion except that those students don't perform better than those not studying with their parents.

This looks normal since we expect of students from that age to be autonomous.

```python
five = pisa_clean[pisa_clean['Study Time With Parent'] < 6]['Study Time With Parent'].value_counts()
six_and_more = pisa_clean[pisa_clean['Study Time With Parent'] >= 6]['Study Time With Parent'].value_counts()
six_and_more.sum()

new_row = pd.DataFrame(data = {
    'Study Time With Parent' : ['6+'],
    'value' : [six_and_more.sum()]
})
df2 = pd.concat([five, new_row['value']], ignore_index=True)
df2

fig1, ax1 = plt.subplots(1, 1, figsize=(10, 10), sharex=True)

patches, texts, autotexts = ax1.pie(df2, 
                                    colors = ['#4c516d','#003152','#0e4d92','#0f52ba','#0080ff','#008ecc','#6593F5'],
                                    labels=['Not working with parent(s)', '1 hour', '2 hours', '3 hours', '4 hours', '5 hours', 'up to 30 hours'], 
                                    autopct='%1.1f%%',
                                    startangle=180
                                   );

for text in texts:
    text.set_color('grey')    
for text in autotexts:
    text.set_color('white')
ax1.axis('equal');

plt.suptitle('How much time students spend with their parent(s)?')
plt.show();
```

```python
scores = pisa_clean.groupby(['Study Time With Parent']).mean().reset_index()[['Study Time With Parent','Overall Score','Math Score','Reading Score','Science Score']]
scores.rename({'Overall Score':'Overall', 'Math Score':'Math', 'Reading Score':'Reading', 'Science Score':'Science'}, axis=1, inplace=True)
data = pd.melt(scores, id_vars=['Study Time With Parent'], 
               value_vars=['Overall','Math','Reading','Science'], 
               var_name="Score Type", value_name='Score')

fig, ax = plt.subplots(1, 1, figsize=(17, 10))

sb.lineplot(ax=ax, data = data, 
            x = 'Study Time With Parent', y='Score', hue='Score Type',
           );
ax.set(xlabel='Nr. of hours', ylabel='Score')
ax.xaxis.labelpad = 20
ax.yaxis.labelpad = 10

plt.suptitle("Time spend studying with parents, does it worth the investment?", y=1.02);
```

## Few words on the single parent?

Scores for single parent and two-parent families are equally distributed but the scores of students living in a single parent family are significanlty less (by approximatly 20 points) than those living in a two-parent family.

Surprinsingly things are getting worse when it comes to native students than non native ones.

From a previous observation, we found out that studying with parents was not efficient and this is worse for single family. An advice for single parent is try as much as much to work on the autonomy of your child.

Finally, for parent who got a upper secondary education or beyond, the parent education level is not so important in a single parent family than it is for a two-parent family.

```python
data = pd.melt(pisa_clean, id_vars=['Student Id', 'Gender', 'Country', 'Study Time With Parent', 'Parents Highest Education', 'Immigration Status', 'Parents Presence'], 
               value_vars=['Overall Score','Math Score','Reading Score','Science Score'], 
               var_name="Score Type", value_name='Score')
data['Score Type'] = data['Score Type'].replace('Overall Score', 'Overall').replace('Math Score','Math').replace('Reading Score','Reading').replace('Science Score','Science')

palette = sb.color_palette("Paired")

plt.figure(figsize = [14, 10])
ax = sb.violinplot(x = 'Score', y='Score Type', hue='Parents Presence',
                   data=data, palette=palette, split=True,
                   scale="count", inner="quartile");
ax.xaxis.labelpad = 20
ax.yaxis.labelpad = 10

x_ticks = np.arange(0, 1000 + 1, 100)
plt.xticks(x_ticks, x_ticks);
plt.suptitle("Is studying harder for the students coming from single parent family?");

```
